<h1> VERİ FİLTRELEME </h1>  

Baba filminin ımdp puanını bulalım. Veri setindekini yerini biliyorsak indexi ile ulaşabilirz.

```python
df['IMDB_Rating'][1]
```

Ancak veri seti çok büyük ve nerede olduğunu bilmiyoruz bunu nasıl bulacağız?

```python
df.loc[df['Series_Title']=="The Godfather"]
```

loc yani location godfather isimli filmin lokasyonunu bulup getiriyor.

Mantığı: df.loc yazdığımız zaman bize dönen çıktının bir tipi var. Bu bir dataframe olarak dönüyor aslında. Bu yüzden sadece bir sütununa erişmek istiyorsak köşeli parantez ile belirtmemiz yeterli olacak.

```python
type(df.loc[df['Series_Title']=="The Godfather"])
```

```python
df.loc[df['Series_Title']=="The Godfather"]['IMDB_Rating']
```

IMDB rating değeri 8 üzeri ve oyları 100000'den fazla olan filmleri getirmesini istersek

```python
df.loc[df['IMDB_Rating']>=8 & (df['No_of_Votes']>=100000)]
```

IMDB rating'i 8in üstünde ve Gross değeri (yani gişe hasılatı) 3000000dan fazla olan filmleri görelim

```python
df.loc[df['IMDB_Rating']>=8 & (df['Gross']>=3000000)]
```

Hata veriyor çünkü tipi int olmayan bir veriye aritmetik işlem(>=) yapmaya çalışıyoruz bunu düzeltmek için veri setiyle oynamamız gerekiyor.

```python
df.dtypes
```

Burada Grossun object olduğunu görüyoruz. Onjectin string olduğunu anlayabiliyor python ama float olduğu zaman veriler sıkıntı oluyor.

```python
#ilk olarak Gross verisi virgüllü bu virgüllerden kurtuluyotuz 
df['Gross']=df['Gross'].str.replace(",", "")
#Gross sütununu numeric tipe çevirmek için to_numeric kullanıyoruz
df["Gross"]=pd.to_numeric(df["Gross"])
type(df["Gross"][0])
```

```python
df.loc[(df['IMDB_Rating']>=8.0) & (df['Gross']>=500000000)]
```

<h3> Filtreleme </h3>

```python
df_filtered=df[df["IMDB_Rating"]>=9.2]
df_filtered
```

```python
#veya...
df_filtered=df.query('IMDB_Rating>=9')
df_filtered
```

```python
f=lambda x: (x["Runtime"].split())[0] #Runtime 140 min gibi iki tane stringten oluştuğu için onu boşluğa göre ayırıp 0. indistekini alıyor
df["RuntimeYeni"]=df.apply(f,axis=1)
```

```python
df.head()
```

```python
df["RuntimeYeni2"]=df["RuntimeYeni"].astype('int')

df_filtered=df.query("RuntimeYeni2>=140")
df_filtered
```

```python
df=df.drop(['RuntimeYeni'],axis=1)
```
